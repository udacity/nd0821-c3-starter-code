```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/census.csv")
df
```

```python
df.info
```

```python
df.isna().sum(axis=0)
```

```python
categorical_features_indices = np.where(df.dtypes == object)[0]
cat_feat= df.columns[categorical_features_indices].tolist()
cat_feat
```

```python
for c in cat_feat:
    print(c)
    print(df[c].value_counts())
    print()
```

```python
df['salary'].value_counts().plot(kind='bar', title='Dataset distribution by class')
```

```python
df.columns = [c.strip() for c in df.columns]
```

```python
df.info()
```

```python
for c in [col.strip() for col in cat_feat]:
    df[c] = df[c].str.strip()
```

```python
df
```

```python
df.to_csv("./data/census.csv", index=False)
```
